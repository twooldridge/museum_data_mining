# Read in and combine records from vertnet and arctos datasets

## Vertnet
This dataset comes from an email correspondence with John Wieczorek from Vertnet. As of early 2018, attempting to download all _P. man_ records from Vertnet only delivered a partial dataset. I had to get the dataset directly emailed, which includes 200,000+ records. This dataset is a snapshot of _P.man._ occurences as of 11-23-2017, hence the date string in the file path. I find it unlikely that things have changed substantially since then, but it might be worth updating.

```r
vertnet = fread("rawdata/Peromyscus_maniculatus20171123.csv.gz",fill=TRUE,sep=",",colClasses = c("character"))

## Create ids ##
vertnet %<>% mutate(ids = paste(institutioncode,catalognumber,sep=":"))

## Select relevant columns. This will become the template for all subsequent datasets ##
vertnet %<>% dplyr::select(ids,
                          scientificname,
                          sex,
                          stateprovince,
                          locality,
                          decimallongitude,
                          decimallatitude,
                          preparations,
                          lifestage,
                          hastissue,
                          dynamicproperties,
                          habitat,
                          year)
```

## Arctos
After talking with the Vertnet people and getting the full Peromyscus dataset, it seems that I can be reasonably confident that all Arctos vertebrate records are included on Vertnet. However, I've noticed a few inconsistencies, mainly that some attributes of Arctos records weren't lifted over to Vertnet. This becomes apparaent for a handful of records that I reference further below in this doc

```r
## fread was throwing annoying erros, this read.csv doesn't take too long
arctos <- read.csv("rawdata/arctos_pmanic_full.csv") 

## Select same columns as vertnet, plus a bit more 
arctos %<>% dplyr::select(GUID,
                          SPECIES,
                          SEX,
                          STATE_PROV,
                          SPEC_LOCALITY,
                          DEC_LONG,
                          DEC_LAT,
                          PARTS,
                          AGE_CLASS,
                          HAS_TISSUES,
                          HABITAT,
                          YEAR,
                          # other fields that would be under 'attributes' in VERTNET
                          TOTAL_LENGTH,
                          TAIL_LENGTH,
                          HIND_FOOT_WITH_CLAW,
                          EAR_FROM_NOTCH,
                          WEIGHT) %>%
    set_colnames(c("ids","scientificname","sex","stateprovince","locality","decimallongitude","decimallatitude",
                   "preparations","lifestage","hastissue","habitat","year",
                   # Morphological traits
                   "total","tail","foot","ear","weight")) %>%
    mutate(dynamicproperties=NA,ids=gsub("*:(\\w+):*",":",ids)) %>%
    mutate_at(.vars=vars(total,tail,foot,ear,weight),.funs=funs(suppressWarnings(as.numeric(str_replace_all(.,"mm|g|\\s+","")))))
```

Now let's combine the datasets, retaining only ARCTOS entries that ARE NOT in vertnet

```r
full_dataset = rbind(vertnet %>%
                     mutate(total=NA,tail=NA,foot=NA,ear=NA,weight=NA),
                     arctos %>% 
                     dplyr::filter(!ids %in% vertnet$ids))
# Change all empty entries to 'NA' to make downstream stuff easier, 
# also paste ';' to end of dynamic properties fields to make REGEX work
full_dataset %<>% 
    mutate_all(.funs = funs(ifelse(.=='',NA,.))) %>% 
    mutate(body=NA,
           dynamicproperties=ifelse(is.na(dynamicproperties),NA,paste0(dynamicproperties,";")))
```

# Extracting morphological data

The morphological data is formattted in a bunch of different ways, so I will need to run a few different commands to squeeze out every last bit of info

__First__, let's get the easy values, where a string (e.g. "tail") is followed directly by a num

```r
## DEFINE FUNCTIONS ## 
# Get first number in a string 
numextract <- function(string){ 
  str_extract(string, "\\-*\\d+\\.*\\d*")
} 

# Gets first number appearing after pattern, given input string & pattern
get_following_num = function(string,pattern){
    expr = paste0("(",pattern,")(.*?)[;,]")
    val = numextract(str_extract(string=string, pattern=regex(expr,ignore_case = TRUE)))[1]
    return(as.numeric(val))
}

# Vectorize function #
v_get_following_num = Vectorize(get_following_num,"string")

## RUN ##
# Runs in about 3 minutes #
ptm <- proc.time()
full_dataset = 
    full_dataset %>% 
    mutate(total = ifelse(is.na(total),v_get_following_num(dynamicproperties,"total"),total),
           tail = ifelse(is.na(tail),v_get_following_num(dynamicproperties,"tail"),tail),
           foot = ifelse(is.na(foot),v_get_following_num(dynamicproperties,"foot"),foot),
           ear = ifelse(is.na(ear),v_get_following_num(dynamicproperties,"ear"),ear),
           weight = ifelse(is.na(weight),v_get_following_num(dynamicproperties,"weight"),weight))
proc.time() - ptm
```

__Second__, we'll get strings formatted roughly like '125-50-20-20' or '125-50-20'

```r
## FUNCTIONS ##
extract_4 = function(string){
    expr = "([a-zA-Z0-9.\\[\\]]{1,8})(-)([a-zA-Z0-9.\\[\\]]{1,7})(-)[a-zA-Z0-9.\\[\\]]{1,7}(-)([a-zA-Z0-9.\\[\\]]{1,7})"
    valstring = str_extract(string,regex(expr,ignore_case=TRUE))
    if (is.na(valstring)){
        expr = "([a-zA-Z0-9.\\[\\]]{1,8})(-)([a-zA-Z0-9.\\[\\]]{1,7})(-)([a-zA-Z0-9.\\[\\]]{1,7})"
        valstring = str_extract(string,regex(expr,ignore_case=TRUE))
        valstring = paste0(valstring,"-NA")
    }
    valstring = valstring %>% str_replace_all(.,"\\[|\\]","")
    vals = valstring %>% str_split(.,"-") %>% unlist(.) %>% as.numeric(.)
    if (length(vals)<4){
        vals = rep(NA,4)
    }
    names(vals) = c("total","tail","foot","ear")
    return(vals)
}

v_extract_4 = Vectorize(extract_4,"string")

#### RUN ####
# Runs in about 10 min, even with parallel cores=4. Not sure how I could speed this up
cl <- parallel::makeCluster(4)
doParallel::registerDoParallel(cl)
ptm <- proc.time()
lst = foreach(i=1:nrow(full_dataset),.packages=c("stringr")) %dopar% {
    row = full_dataset[i,]
    if (anyNA(row[,c("total","tail","foot","ear")])){
        traits = suppressWarnings(extract_4(row$dynamicproperties))
        for (trait in names(traits)){
            if (is.na(row[,trait])){row[,trait] = traits[trait]}
        }
    }
    return(row)
}
proc.time() - ptm
parallel::stopCluster(cl)
full_dataset = rbindlist(lst) ## Binding after the for loop, rather than in the foreach call, is MUCH faster
```

__Third__, let's get the special cases that previous operations didn't pick up

```r
#### FUNCTIONS ####
# If total length was just indicated by "length"
get_length = function(string){
    val = str_extract(string,regex('(?i)([;,]\\s+length=)(.*$)',ignore_case=TRUE))
    val = numextract(val)
    return(as.numeric(val))
}

v_get_length = Vectorize(get_length)

# weight was indicated like '=18g'
capture_trailing_weight = function(string){
    val = str_extract(string,regex('([=])\\w+[g]',ignore_case=TRUE)) %>% str_replace_all(.,"g|=","")
    val = suppressWarnings(as.numeric(val))
    return(val)
}

v_capture_trailing_weight = Vectorize(capture_trailing_weight)


#### RUN ####
full_dataset %<>% mutate(body=NA)
full_dataset = 
    full_dataset %>% 
        mutate(total = ifelse(is.na(total),v_get_length(dynamicproperties),total),
               weight = ifelse(is.na(weight),v_capture_trailing_weight(dynamicproperties),weight),
               weight = ifelse(is.na(weight),v_get_following_num(dynamicproperties,"mass"),weight),
               total = ifelse(is.na(total),v_get_following_num(dynamicproperties,"headbodylength"),total))
```

There are a handful of samples for which arctos had morphological data, but vertnet had an empty dynamic properties field. Because vertnet sample IDs were prioritized in the merge, that data was lost. Now we'll try to recover that:

```r
arctos_has_data = arctos %>% filter(total!="") %>% .$ids
full_missing_data = full_dataset %>% filter(is.na(total)) %>% filter(ids %in% arctos_has_data) %>% .$ids
for (id in full_missing_data){
    idx = grep(id,full_dataset$ids)
    if (length(idx) == 1){
        full_dataset[idx,c("total","tail","foot","ear","weight")] = arctos %>% filter(ids == id) %>% dplyr::select(total,tail,foot,ear,weight)
    }   
}
```

Now let's explore what types of records ended up having no data, and see if we can pick up any remaining cases

**So a full-pop sequencing approach will range from 200-300 specimens.** This is a very significant range for WGS, not as meaningful for DDRAD. What would happen if we did 3 specimens from each locale?

```r
loaned = loans %>% filter(Requested == "Y",!is.na(Group_ID)) %>% pull(Group_ID) %>% unique %>% length * 3
in_house = loans %>% filter(is.na(Group_ID),!ID %in% c("P. m. nubiterrae","P. m. bairdii","Siuslaw mice","Baker City Mice")) %>% nrow * 3
print(paste0("Target estimate: ",loaned + in_house))

loaned = loans %>% filter(Received == "Y",!is.na(Group_ID)) %>% pull(Group_ID) %>% unique %>% length * 3
in_house = loans %>% filter(is.na(Group_ID),!ID %in% c("P. m. nubiterrae","P. m. bairdii","Siuslaw mice","Baker City Mice")) %>% nrow * 3
print(paste0("Low estimate: ",loaned + in_house))

loaned = loans %>% filter(Requested == "Y",!is.na(Group_ID)) %>% pull(Group_ID) %>% unique %>% length * 3
in_house = loans %>% filter(is.na(Group_ID),!ID %in% c("P. m. nubiterrae","P. m. bairdii","Siuslaw mice","Baker City Mice")) %>% nrow * 3
print(paste0("High estimate: ",loaned + in_house + 4*3))
```

** In this case, a full-pop sequencing approach would range from 65-85 specimens.** This would be much more reasonable for WGS, but what would you be able to do with this sort of sparse population samplilng. Of course, it's important to consider the *likely* case where we don't sequence every population, but just a handful of the interesting ones

Let's do some rough calculations based on library prep and WGS costs, assuming 20X or 10X coverage and optimal NovaSeq S4 output (500GB)

```r
WGS_prep_per_sample = 32
WGS_10X_per_sample = 6200/(500/(2.51 * 10)) # Lane cost divided by samples per lane
WGS_20X_per_sample = 6200/(500/(2.51 * 20))
WGS_5X_per_sample = 6200/(500/(2.51 * 5))
#DDRAD_prep_per_sample =
#DDRAD_seq_per_sample =

strategy_df = data.frame(sampling_strategy =c(rep("subset",3),rep("whole",3)),
                         number_pops = c(rep("low",2),rep("target",2),rep("high",2)),
                         sample_num = c(66,75,87,213,248,288))
strategy_df %>% mutate(WGS_prep = sample_num*WGS_prep_per_sample,
                      WGS_5X = sample_num*WGS_5X_per_sample,
                      WGS_10X = sample_num*WGS_10X_per_sample,
                      WGS_120X = sample_num*WGS_20X_per_sample)
```
